```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2 * np.pi, 200)
omega1 = np.sqrt(1011)
omega2 = np.sqrt(7)
```

```python
f1 = np.sin(omega1 * t)
f2 = np.sin(omega2 * t)
f12 = f1 + f2
```

```python
plt.plot(t, f12)
plt.show()
```

```python
omega3 = 1011
omega4 = 2
f3 = np.sin(omega3 * t)
f4 = np.sin(omega4 * t)
f34 = f3 + f4
```

```python
plt.plot(t, f34)
plt.show()
```
